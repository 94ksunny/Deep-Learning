```python
def outlier_treat(x,val,how):
    if how=='upper':
        x=x.clip_upper(x.quantile(val))
    elif how=='lower':
        x=x.clip_lower(x.quantile(val))
    else:
        print('Enter the valid value')
    return x
```

```python
tr2['age_building']=tr2[['age_building']].apply(lambda v:outlier_treat(v,0.01,'lower'))
```

```python
col=['age_building','plinth_area_sq_ft','height_ft_pre_eq','height_ft_post_eq']
```

```python
tr2[col]=tr2[col].apply(lambda v:outlier_treat(v,0.99,'upper'))
```

```python
mapping={'Grade 1':0,'Grade 2':1,'Grade 3':2,'Grade 4':3,'Grade 5':4}
inv_mapping={j:i for i,j in mapping.items()}
```

```python
tr3=tr2.copy()
```

```python
tr3['damage_grade']=tr3['damage_grade'].apply(lambda v:mapping[v])
```

```python
target=tr3['damage_grade']
tr3.drop('damage_grade',axis=1,inplace=True)
```

```python
num=tr3.select_dtypes(include=['float64','int64']).columns
cat=tr3.select_dtypes(include='object').columns
```

```python
print('Categorical Columns are: ',cat)
print('Numerical Columns are: ',num )
```

```python
def get_dummy(df,col):
    cat_df=pd.get_dummies(df[col],drop_first=True)
    df=pd.concat([cat_df,df],axis=1)
    df.drop(col,axis=1,inplace=True)
    return df
```

```python
tr3=get_dummy(tr3,cat)
```

```python
tr3=tr3.astype(float)
tr3.info()
tr3.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
scaled_tr3=sc.fit_transform(tr3)
```

```python
train_X,test_X,y_train,y_test=train_test_split(scaled_tr3,target,stratify=target,test_size=0.1,random_state=42)
```

```python
print(train_X.shape)
print(test_X.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
### Build Neural Network Model
```

```python
from keras.models import Sequential
from keras.layers import Dense,Activation,Dropout
from keras.utils import to_categorical
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
y_train_cat=to_categorical(y_train)
y_test_cat=to_categorical(y_test)
```

```python
def baseline_model():
    #create model
    model=Sequential()
    model.add(Dense(512,input_dim=79,activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(512,activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(256,activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(128,activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(5,activation='softmax'))
    #compile
    model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
    return model
```

```python
model=Sequential()
model.add(Dense(512,input_dim=79,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(512,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(256,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(5,activation='softmax'))
#compile
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history=model.fit(train_X,y_train_cat,batch_size=256,epochs=10,verbose=1,validation_data=[test_X,y_test_cat])
```

```python
scores = model.evaluate(test_X, y_test_cat, verbose=1)
```
